```python
import json
import pandas as pd
import numpy as np
import csv

data = '[{ "Gender": "Male", "HeightCm": 171, "WeightKg": 96 } , { "Gender": "Male", "HeightCm": 161, "WeightKg": 85 } , { "Gender": "Male", "HeightCm": 180, "WeightKg": 77 } , { "Gender": "Female", "HeightCm": 166, "WeightKg": 62} , { "Gender": "Female", "HeightCm": 150, "WeightKg": 70} , { "Gender": "Female", "HeightCm": 167, "WeightKg": 82}]'
data_dict = json.loads(data)
print(data_dict)
```

```python
df = pd.DataFrame(data_dict)
```

```python
df
```

Calculate the BMI (Body Mass Index), BMI Category and Health risk from Table of the person and add them as 3 new columns

```python
df['HeightM'] = df['HeightCm'].apply(lambda x: x*0.01)
```

```python
df['BMI'] = df['WeightKg']/df['HeightM'] #.assign(lambda x : (x['WeightKg'] / x['HeightM']))
```

```python
df
```

```python
def BMI_CATEGORY(BMI):
    if BMI <= 18.4 :
        return('Underweight')
    elif BMI >=18.5 and BMI <= 24.9:
        return('Normal weight')
    elif BMI >=25 and BMI <= 29.9:
        return('Overweight')        
    elif BMI >=30 and BMI <= 34.9:
        return('Moderately obese')
    elif BMI >= 35 and BMI <= 39.9:
        return('Severely obese')
    elif BMI > 40:
        return('Very severely obese')
```

```python
def HEALTH_RISK_CHECK(BMI):
    if BMI <= 18.4 :
        return('Malnutrition risk')
    elif BMI >=18.5 and BMI <= 24.9:
        return('Low risk')
    elif BMI >=25 and BMI <= 29.9:
        return('Enhanced risk')        
    elif BMI >=30 and BMI <= 34.9:
        return('Medium risk')
    elif BMI >= 35 and BMI <= 39.9:
        return('High risk')
    elif BMI > 40:
        return('Very high risk')
```

```python
df['BMI_Category'] = df['BMI'].apply(BMI_CATEGORY) 
```

```python
df['Health_Risk'] = df['BMI'].apply(HEALTH_RISK_CHECK) 
```
